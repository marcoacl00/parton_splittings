```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
import matplotlib.ticker as ticker

from matplotlib import rcParams
#font_name = 'Times New Roman'

rcParams['font.size'] = 8
rcParams['mathtext.fontset'] = 'custom'
```

```python
class OOMFormatter(ticker.ScalarFormatter):
    def __init__(self, order=0, fformat="%1.1f", offset=True, mathText=True):
        self.oom = order
        self.fformat = fformat
        ticker.ScalarFormatter.__init__(self,useOffset=offset,useMathText=mathText)
    def _set_order_of_magnitude(self):
        self.orderOfMagnitude = self.oom
    def _set_format(self, vmin=None, vmax=None):
        self.format = self.fformat
        if self._useMathText:
            self.format = r'$\mathdefault{%s}$' % self.format
```

```python
time = np.loadtxt("F_time_largeNc.txt")[:, 0]
F_values = np.loadtxt("F_time_largeNc.txt")[:, 1]

time2 = np.loadtxt("F_time_Nc.txt")[:, 0]
F_values2 = np.loadtxt("F_time_Nc.txt")[:, 1]

```

```python
fig, ax1 = plt.subplots(figsize=(6,4))
plt.title(r"$E = 10$ GeV, $\theta$ = 0.5, $z$ = 0.5, L = 2 fm")
ax1.plot(time, F_values, color = "red", label = "N = 40 large Nc", linewidth = 1, linestyle = "--")
ax1.plot(time2, F_values2, color = "black", label = "N = 32 Finite Nc", linewidth = 1)

ax1.set(xlabel = "time [fm]", ylabel = "$F_{med}^{in-in}$")

ax1.grid(True)
ax1.grid(color='0.95', linestyle='--', linewidth=1)

#ax1.xaxis.set_major_locator(MultipleLocator(0.25))

ax1.legend( prop={'size': 9})

plt.show()
```
